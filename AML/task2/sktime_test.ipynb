```python
import numpy as np
import pandas as pd
from sktime.classification.interval_based import TimeSeriesForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score
from sktime.datasets import load_arrow_head
from sktime.datatypes import check_is_mtype
from sktime.utils import mlflow_sktime  
```

```python
X_train_orig = pd.read_csv('X_train.csv')
y_train_orig = pd.read_csv('y_train.csv')
X_test_orig = pd.read_csv('X_test.csv')
```

```python

label=y_train_orig.iloc[:,1]
feature=X_train_orig.iloc[:,1:].dropna(axis=1).to_numpy()
print((feature))
#valid, error_msg, metadata = check_is_mtype(feature, "univariate", return_metadata=True)
```

```python
X, y = load_arrow_head()
#check_is_mtype(X, 'univariate')
print(type(X))
X_train, X_test, y_train, y_test = train_test_split(feature, label)
classifier = TimeSeriesForestClassifier()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
score = f1_score(y_test, y_pred, average='micro')
score
```

```python
mlflow_sktime.save_model(  
    sktime_model=classifier,
    path='model')  
```
